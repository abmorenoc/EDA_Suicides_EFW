<h1>Comparando la libertad económica de un pais con su número de suicidios </h1>

En esta libreta se pretende realizar un análisis exploratorio de dos fuentes de datos diferentes y ver cómo se relacionan entre ellas. Una es una dataset que contiene información del número de suicidios por año de 141 países, desde el año 1979 hasta el 2016. El otro dataset contiene datos de la libertad economica de 162 paises, desde el año 1970 hasta el 2016.


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Primero leemos los datos que provienen del dataset del número de suicidios por año.

```python
df = pd.read_csv("who_suicide_statistics.csv")
df.head()
```

<h3>Explorando el dataset de suicidios por año</h3>

Primero exploremos los datos con los que vamos a estar trabajando. 

```python
print('Volumen:',df.shape)
```

```python
df.describe()
```

Numero de elementos nulos por columna


```python
df.isnull().sum()
```

<h3> Limpieza de datos </h3>

Las columnas "suicides_no" y "population" tiene un número significativo de valores nulos por lo tanto los reemplazamos por ceros. No es factible hacer una interpolación linear o imputación de sus medias ya que se hará un agrupamiento por año y país. Por resultante tendremos la suma de los suicidios por año y máximo valor de población en ese año. Si reemplazamos los valores nulos de suicidios por año y la poblacion estaríamos obteniendo un resultado muy alejado de la realidad.

```python
# df = df.sort_values(['country', 'year'])
# df.interpolate(method='linear', inplace=True)
# df.isnull().sum()
```

```python
df.update(df[['suicides_no','population']].fillna(0))
suicides = df.groupby(['country','year']).agg({'suicides_no':'sum', 'population':'max'}).reset_index()
```

Agregaremos una columna extra con el porcentaje de suicidios en contraste a la población del país.

```python
suicides['percentage_suicides'] = (suicides.suicides_no * 100)/suicides.population 
suicides = suicides.replace([np.inf, -np.inf], 0)
suicides.head()
```

<h3>Análisis exploratorio de los datos</h3>

Ya tenemos un dataset sin valores nulos, ahora utilizaremos algunas gráficas para obtener información relevante de este conjunto de datos

```python
top_suicides_2015 = suicides[suicides.year==2015].sort_values(by='suicides_no', ascending=False).head(15)

sns.set_palette("Set1", 8, .75)
sns.set_style('whitegrid')

top_suicides_2015.plot('country', 'suicides_no', kind='bar', figsize=(14,8), rot=45)

plt.xlabel('Pais')
plt.ylabel('Numero de siucidios')
plt.title('Los 15 paises con mas numero de suicidios en 2015')
```

```python
top_suicides_2015 = suicides[suicides.year==2007].sort_values(by='percentage_suicides', ascending=False).head(15)

sns.set_palette(sns.color_palette("husl", 8))
sns.set_style('whitegrid')

top_suicides_2015.plot('country', 'percentage_suicides', kind='bar', figsize=(14,8), rot=45)

plt.xlabel('Pais')
plt.ylabel('Porcentaje de siucidios')
plt.title('Los 15 paises con mas porcentaje de suicidios en 2015')
```

```python
sns.set_palette(sns.color_palette("colorblind",15))
sns.set_style('whitegrid')

fig, ax = plt.subplots()

names = top_suicides_2015['country'].tolist()
for name in names:
    ndf = suicides.loc[suicides['country']==name]
    ax = ndf.plot(ax=ax, kind='line', x='year', y='percentage_suicides', label=name, figsize=(18,10), linewidth=2.5)

plt.xlim((1985, 2015))
plt.xlabel('Año')
plt.ylabel('Suicidios')
plt.title('Suicidos en el mundo 1985 - 2015')
```

```python
top_percentage_suicides_2015 = suicides[suicides.year==2015].sort_values(by='percentage_suicides', ascending=False).head(15)
top_percentage_suicides_2015

sns.set_palette(sns.color_palette("colorblind",15))
sns.set_style('whitegrid')

fig, ax = plt.subplots()

names = top_percentage_suicides_2015['country'].tolist()
for name in names:
    ndf = suicides.loc[suicides['country']==name]
    ax = ndf.plot(ax=ax, kind='line', x='year', y='percentage_suicides', label=name, figsize=(18,10), linewidth=2.5)

plt.xlim((1985, 2015))
plt.xlabel('Año')
plt.ylabel('Porcentaje de suicidios')
plt.title('Porcentaje de suicidios en el mundo 1985 - 2015')
```

<h3>Explorando el dataset de la libertad economica mundial</h3>

El dataset de "The Economic Freedom of the World" creado por Fraser Institute es la principal medida mundial de libertad economica por pais. Mide el nivel en que las politicas y instituciones del pais apoyan a la libertad economica.

```python
df2 = pd.read_csv("efw_cc.csv")
# df2 = df2.sort_values(['countries','year'])
df2.head(10)
```

```python
df2.describe()
```

```python
df_clean = df2
df2.isnull().sum()
```

<h3>Limpiando los datos de The Economic Freedom of the World</h3>

Primero eliminaremos la columnas que tengan mas de 1/3 de valores nulos. Ya que realizaremos una interpolacion linear para imputar los datos faltantes, si lo hacemos en una columna con muy pocos datos el resultado estaria muy alejado de la realidad.

```python
# Eliminando las columnas con mas de 1242 valores nulos

df_clean = df2.loc[:, (df2.isnull().sum(axis=0) <= 1242)]
df_clean.interpolate(method='linear', inplace=True)
df_clean.isnull().sum()
```

Una vez realizada la interpolacion lineal quedaron muy pocos valores faltantes, asi que aplicamos dropna() para terminar de eliminar todos los renglones con datos faltantes restantes.

```python
df2 = df2.dropna()
df2.head()
```

```python
top_efw_2015 = df2[df2.year==2015].sort_values(by='ECONOMIC FREEDOM', ascending=False).head(15)

sns.set_palette(sns.color_palette("colorblind",15))
sns.set_style('whitegrid')

fig, ax = plt.subplots()

names = top_efw_2015['countries'].tolist()
for name in names:
    ndf = df2.loc[df2['countries']==name]
    ax = ndf.plot(ax=ax, kind='line', x='year', y='ECONOMIC FREEDOM', label=name, figsize=(18,10), linewidth=2.5)

plt.xlim((1995, 2015))
plt.xlabel('Año')
plt.ylabel('Libertad Economica')
plt.title('Libertad economica en el mundo 1995 - 2015')
```

<h3>Seleccionando variables</h3>

Seleccionaremos el subconjunto de variables utilizando el metodo de filtrado. Este consiste en obtener las columnas que tengan mayor coorrelacion con la variable objetivo, que en este caso es "Economic Freedom". 

```python
data_num = df2._get_numeric_data()
data_cor = data_num.corr()

sns.set(font_scale=1.4)

plt.figure(figsize=(13,13))
sns.heatmap(data_cor,  square=True, cmap='coolwarm_r')
```

```python
#Correlation with output variable
cor_target = abs(data_cor["ECONOMIC FREEDOM"])

relevant_features = cor_target[cor_target>0.6]
relevant_features
```

```python
# Obteniendo las columnas con caracteristicas principales
efw = df2.loc[:,['year','countries','ECONOMIC FREEDOM', 'quartile', '2_property_rights', '3d_freedom_own_foreign_currency',
                 '3_sound_money','4d_control_movement_capital_ppl','4_trade','5a_credit_market_reg', '5_regulation']]

efw.rename(columns={"countries": "country",
                     "ECONOMIC FREEDOM": "economic_freedom",
                     "2_property_rights": "property_rights",
                     "3d_freedom_own_foreign_currency": "foreign_currency",
                     "3_sound_money": "sound_money",
                     "4d_control_movement_capital_ppl": "control_movement",
                     "4_trade": "trade",
                     "5a_credit_market_reg": "credit_market_reg",
                     "5_regulation": "regulation"}, inplace=True)
efw.head(10)
```

```python
data_num = efw._get_numeric_data()
data_cor = data_num.corr()

sns.set(font_scale=1.4)

plt.figure(figsize=(13,13))
sns.heatmap(data_cor,  square=True, cmap='coolwarm_r')
```

<h2>Fusionando los datos</h2>


```python
# Merging Data Frames
df_final = efw.merge(suicides, how="left", on=['country','year'])

# These datasets dont have the same countries and year so we just remove the NaN

df_final = df_final.dropna()
df_final.head()
```

```python
# Guardamos el dataset final
# df_final.to_csv('efw_suicides_rate.csv',index = False, header = True)
```

```python
fig = plt.gcf()
fig.set_size_inches(16, 10)
sns.set(font_scale=1.4)

df_final.quartile = df_final.quartile.astype('int64')

sns.scatterplot(x='economic_freedom', y='suicides_no', data=df_final, s=45,\
                hue='quartile', palette=["#9b59b6", "#3498db", "#e74c3c", "#2ecc71"])
plt.xlabel('Libertad Economica')
plt.ylabel('Numero de suicidios')
plt.title('Relacion entre libertad economica y suicidios')
```

```python
data_num = df_final._get_numeric_data()
data_cor = data_num.corr()

sns.set(font_scale=1.4)

plt.figure(figsize=(13,13))
sns.heatmap(data_cor,  square=True, cmap='BuPu')
```
